```python
# נתיב החסד- סופר חסד בע"מ -> for store name
# יינות ביתן -> from store name
# זול ובגדול בע"מ -> no information
#  פז חברת נפט בע"מ -> store name
# אלמשהדאוי קינג סטור בע"מ' -> no inforamtion
# Dabach -> no inforamtion
# 'פרש מרקט' -> store name
# מ. יוחננוף ובניו -> store name 

#data_frame = data_frame.drop(columns=['zipcode'])
def fix_address(x):
    address_can_be_found_in_name = [ 'יינות ביתן', 'פרש מרקט', 'מ. יוחננוף ובניו', 'פז חברת נפט בע"מ', 'נתיב החסד- סופר חסד בע"מ']
    if x['city'] != 'unknown' and x['address'] != "unknown":
        return x['city'] + " , " + x['address']
    if x['city'] == 'unknown' and x['chainname'] in address_can_be_found_in_name:
        return x['storename']
    return x['city']

stores['location'] = stores.apply(fix_address,axis=1)
```
