```python
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

with_number = prices[prices['unitqty'].apply(has_numbers)][['unitqty','unitofmeasure']]
(with_number['unitqty'] == with_number['unitofmeasure']).all() # we can remove letters from unitofmeasure and numbers from unitqty
```

```python
def collect_details(x):
    return pd.Series([x['itemname'].unique(), x['chainid'].unique()])

itemcode_to_itemname = prices.groupby(['itemcode']).apply(collect_details)
itemcode_to_itemname.columns = ['names','chanin_selling']
```

```python

itemcode_to_itemname[itemcode_to_itemname['names'].apply(lambda x:len(x)>1)].to_csv("itemcode_as_multiple_name.csv",index=False)

itemcode_to_itemname[itemcode_to_itemname['chanin_selling'].apply(lambda x:len(x) == 1)].to_csv("itemcode_selling_only_once.csv",index=False)
```
