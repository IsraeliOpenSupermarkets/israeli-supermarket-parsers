```python
def what_field_has_changed(data,with_values=True):
    #print("!")
    if data.shape[0] == 1:
        return pd.Series()
    else:
        values = []
        keys = []
        for c in data.columns:
            uniqu = data[c].unique().tolist()
            if len(uniqu) > 1:
                values.append(data[c].unique().tolist())
                keys.append(c)
        if with_values:
            return pd.Series(data=dict(zip(keys,values)))
        else:
            return keys
            
changes = prices[prices.same_itemcodes.apply(len) > 1].drop(columns=['same_itemcodes']).groupby(['clean_itemname','file_id']).apply(what_field_has_changed)
changes = changes.reset_index().rename(columns={'level_2':"field",0:"changes"})
changes
```

```python
# compute what field has changed.
what_field_changed = changes.groupby(['clean_itemname','file_id']).apply(lambda x:x.field.values).reset_index(name="field changed")
what_field_changed
```

```python
reasoning = {
    "different_manufacture" : ["manufacturername","manufacturername","manufactureritemdescription","manufacturecountry"],
    "different_packing" : ["unitqty","quantity","unitofmeasure","qtyinpackage"],
    "different_pricing_method" : ["unitofmeasureprice","bisweighted"],
    "different_discount_method" : ['allowdiscount'],
}

# assuming those are different prodcut, the reasons above should explain the change in the ones below.
field_that_should = ['priceupdatedate',"itemprice"]

def fields_that_explains_well_duplicate(x):
    result = set()
    for field in x:
        
        #
        to_add = None
        for reason,values in reasoning.items():
            if not to_add and field in values:
                to_add = reason
        
        if not to_add:
            to_add = field
        
        if field not in field_that_should:
            result.add(to_add)
            
    return sorted(list(result))

what_field_changed['field changed group'] = what_field_changed['field changed'].apply(fields_that_explains_well_duplicate)
what_field_changed['field changed group']
```

```python
what_field_changed['no reason'] = what_field_changed['field changed group'].apply(lambda x:len(set(reasoning.keys()) & set(x)) == 0) 
no_reason = what_field_changed[what_field_changed['no reason']].merge(changes)
no_reason
```

```python
import difflib


def string_diff(a,b):
    changes = set()
    print('{} => {}'.format(a,b))  
    for i,s in enumerate(difflib.ndiff(a, b)):
        if s[0]==' ': continue
        elif s[0]=='-':
            print(u'Delete "{}" from position {}'.format(s[-1],i))
        elif s[0]=='+':
            print(u'Add "{}" to position {}'.format(s[-1],i))
        changes.add(s[-1])
    print()
    
    return changes
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=1)
tfidf_matrix = tf.fit_transform(prices['clean_itemname'])


tfidf_matrix.shape[0] == len(prices['clean_itemname']) and tfidf_matrix.shape[1] == len(tf.vocabulary_)
```

```python
# high sim score with different itemcode
from sklearn.metrics.pairwise import cosine_similarity
sim_tfidf_matrix = cosine_similarity(tfidf_matrix)
```

```python
import numpy as np
matchs = np.argwhere(np.tril(sim_tfidf_matrix,-1) > 0.9)
```

```python
product1 = []
product2 = []
score = []
for m1,m2 in matchs:

    row_1 = prices.iloc[m1]
    row_2 = prices.iloc[m2]

    if row_1['itemcode'] != row_2['itemcode']:
        product1.append(row_1[['itemname','chainid','itemcode']].values)
        product2.append(row_2[['itemname','chainid','itemcode']].values)
        score.append(sim_tfidf_matrix[m1,m2])

high_sim_diffrent_itemcode = pd.DataFrame(np.concatenate((np.array(product1),np.array(product2)),axis=1),
                                            columns=['product_1_itemname',"product_1_chainid","product_1_itemcode",
                                                     'product_2_itemname',"product_2_chainid","product_2_itemcode"]
                  )
high_sim_diffrent_itemcode['score'] = score
high_sim_diffrent_itemcode.drop_duplicates().to_csv("high_sim_diffrent_itemcode.csv")
```
