```python
import pandas as pd
df = pd.read_csv("promo_full.csv").drop(columns=['Unnamed: 0'])
df.head()
```

```python
json_parse = ['additionalsminbasketamount','additionalscoupon','additionalstotals','promotionitems','clubs','promotiondetails','additionalrestrictions','giftsitems']

import json
import ast
def parse_json(x):
    if isinstance(x,list) or isinstance(x,dict):
        return x
    try:
        if x in ['not_apply','NO-CONTENT']:
            return []
        # 
        evalute_experssion = eval(str(x))
        return evalute_experssion
        
        #return json.loads(evalute_experssion)
    except (json.JSONDecodeError,TypeError) as e:
        raise ValueError(f"Failed {x} with {e}, {type(evalute_experssion)}")


for column in json_parse:
    print(column)
    df[column] = df[column].apply(parse_json)
```

```python
boolean_parse = ['isgiftitem','allowmultiplediscounts','isweightedpromo']

def parse_boolean(column):
    new_column = column.replace('not_apply',-1).replace('NO-CONTENT',-1).astype(int)
    assert len(set(new_column.unique()) - set([0,1,-1,-2])) == 0
    return new_column
    
for column in boolean_parse:
    print(column)
    df[column] = parse_boolean(df[column])
```

```python
num_parse = ['minnoofitemsoffered','itemtype','minpurchaseamnt','chainid','subchainid','storeid','bikoretno','minnoofitemofered','itemcode','rewardtype','promotionid','discounttype','weightunit','clubid']


def parse_number(column):
    if column.dtype == 'int64':
        return column
    
    assert pd.to_numeric(column,errors='coerce').min() >= 0, "minimal value should be 0"
    contains_dot = column.astype(str).str.contains('\.')
    
    assert ~(contains_dot.any()), f"should not contains float, example= {column[contains_dot].values}"
    
    new_column = column.replace('not_apply',-1).replace("NOT_APPLY",-1).replace('NO-CONTENT',-1).astype(int)
    
    return new_column

for column in num_parse:
    print(column)
    df[column] = parse_number(df[column])
```

```python
floats_parse = ['minpurchaseamount','discountedprice','discountedpricepermida','discountrate','maxqty','minqty','additionalsgiftcount']

def parse_float(column):
    
    new_column = column.replace('not_apply',-1).replace("NOT_APPLY",-1).replace('NO-CONTENT',-1).astype(float)
    
    return new_column

for column in floats_parse:
    print(column)
    df[column] = parse_float(df[column])
```

```python
import numpy as np
time_parse = ['promotionstarthour','promotionendhour']

def parse_time(column):
    if hasattr(column,'dt'):
        return column 
    
    # replace all knowen not time values with 'None'
    column_with_only_time = column.replace('not_apply',None).replace("NOT_APPLY",None).replace('NO-CONTENT',None)
    
    # append ":00" to time ends
    time_parts = column_with_only_time.astype(str).str.split(":").apply(len)
    column_with_only_time.loc[time_parts==2] += ":00"
    
    return pd.to_datetime(column_with_only_time,format='%H:%M:%S')

for column in time_parse:
    print(column)
    df[column] = parse_time(df[column])
```

```python
date_parse = ['priceupdatedate','promotionstartdate','promotionenddate','promotionupdatedate']


def parse_date(column,date_format = '%Y-%m-%d %H:%M:%S',missing=np.nan):
    
    if hasattr(column,'dt'):
        return column 
    
    mid_column = pd.to_datetime(column,format=date_format,errors='coerce')
    row_failed_to_parse = (np.isnat(mid_column)) & (~column.isin(['not_apply','NOT_APPLY','NO-CONTENT']))
    
    assert column[row_failed_to_parse].empty ,f"{column[row_failed_to_parse].shape} found foramt {column[row_failed_to_parse].unique()}"
    
    return pd.to_datetime(column.replace('not_apply',missing).replace("NOT_APPLY",missing).replace('NO-CONTENT',missing),format='%Y-%m-%d')
    
    
for column in date_parse:
    print(column)
    df[column] = parse_date(df[column])
    
```

```python
# couple 
couples = [
    ['promotionstartdate','promotionstarthour'],
    ['promotionenddate','promotionendhour']
]

def combine_datetime(df,date_column,time_column):
    promotionstart_datetime =  df.loc[~df[time_column].isna(),date_column].dt.date.astype(str) + " " + df.loc[~df[time_column].isna(),time_column].dt.time.astype(str)
    df.loc[~df[time_column].isna(),date_column] = pd.to_datetime(promotionstart_datetime)

    
    return df.drop(columns=[time_column]).rename(columns={date_column:f"{date_column}time"})

for date_column,time_column in couples:
    df = combine_datetime(df,date_column,time_column)
    
    time_parse.remove(time_column)
    date_parse.remove(date_column)
    date_parse.append(f"{date_column}time")
```

```python
# 
str_prase = ['file_id','promotiondescription','remarks','remark']

def parse_str(column):
    for col in ['not_apply','NOT_APPLY','NO-CONTENT']:
        column = column.replace(col,'')
    
    return column.astype(str)

for col in str_prase:
    df[col] = parse_str(df[col])
```

```python
assert len(set(df.columns) - set(['promotionenddatetime','promotionstartdatetime']) - set(str_prase+date_parse+time_parse+floats_parse+num_parse+boolean_parse+json_parse)) == 0
```

 # Convert to numeric 

```python
numeric_df = df.copy()

def project_str_to_numeric(df,column_name):
    df[f'{column_name}_len'] = df[column_name].apply(len)
    df[f'{column_name}_all_digits'] = df[column_name].str.count('\d')
    df[f'{column_name}_digits_groups'] = df[column_name].str.findall(r'(\d+(?:\.\d+)?)').apply(len)
    df[f'{column_name}_preceage'] = df[column_name].str.count('%')
    
    return df.drop(columns=column_name)


for col in str_prase:
    numeric_df = project_str_to_numeric(numeric_df,col)
```

```python
def project_json_to_numeric(df,column_name):
    df[f'{column_name}_len'] = df[column_name].apply(len)

    return df.drop(columns=column_name)

for column in json_parse:
    print(column)
    numeric_df = project_json_to_numeric(numeric_df,column)
```
